## Algoritma Naive Bayes dan Preprocessing (pengenalan)

```python
new_df = df.drop(['Unnamed: 0', 'CLIENTNUM', 'Customer_Age', 'Credit_Limit'], axis=1)
new_df
```

```python
# Mendefinisikan fungsi naive_bayes.
def naive_bayes(kategori):
    total = len(new_df)
    
# Menghitung jumlah data pada kolom "Attrition_Flag" berisi 'Existing Customer' dan 'Attrited Customer'.
    target_counts = new_df['Attrition_Flag'].value_counts()
    existing_counts = target_counts['Existing Customer']
    attrited_counts = target_counts['Attrited Customer']

    existing_prob = 1
    attrited_prob = 1

# Melakukan iterasi pada setiap item pada dictionary kategori yang merupakan variabel-variabel yang digunakan sebagai fitur.
    for key, value in kategori.items():
        existing_prob *= (len(new_df[(new_df[key] == value) & (new_df['Attrition_Flag'] == 'Existing Customer')]) / existing_counts)
        attrited_prob *= (len(new_df[(new_df[key] == value) & (new_df['Attrition_Flag'] == 'Attrited Customer')]) / attrited_counts)

# Menghitung nilai probabilitas dari kategori 'Existing Customer' dan 'Attrited Customer' pada fitur tertentu.
    existing_post = existing_prob * (existing_counts / total)
    attrited_post = attrited_prob * (attrited_counts / total)

# Membandingkan nilai probabilitas posterior dari masing-masing kategori dan mengembalikan hasil prediksi dan nilai probabilitas posterior yang lebih tinggi.
    if existing_post > attrited_post:
        return ('Existing Customer', round(existing_post, 3))
    else:
        return ('Attrited Customer', round(attrited_post, 3))
```

```python
tes = naive_bayes({'Gender': 'F', 'Education_Level': 'Post-Graduate', 'Marital_Status' : 'Married', 'Income_Category' : 'Unknown', 'Card_Category' : 'Blue'})
tes
```
